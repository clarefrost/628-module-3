```python
dictory_wkn=
import json
import pandas as pd
def readjson(filepath):
    file = open(filepath, 'r', encoding='utf-8')
    dat={}
    for i,line in enumerate(file,start=1):
        dat_new=json.loads(line)
        if i==1:
            dat=dat_new
            for key in dat :
                dat[key]=[dat[key]]
        else:
            for key in dat:
                dat[key].append(dat_new[key])
    file.close()
    data=pd.DataFrame(dat)
    return data
reviews=readjson("C:/Users/58479/Desktop/628/module3/Stat628/Stat628/review.json")
reviews.head(10)
```

```python
business=readjson("C:/Users/58479/Desktop/628/module3/Stat628/Stat628/business.json")
business.head(10)
```

```python
business.groupby('city')["city"].value_counts().sort_values(ascending=False)[:20]
```

```python
business_Las=business.loc[(business["city"]=="Las Vegas")&(business['categories'].str.contains("Restaurants",na=False))]
business_Las.head(10)
```

```python
business_keep=business_Las.drop(['name','state','postal_code','hours','address'], axis=1)
business_keep.rename(columns={'review_count':'review_count_business','stars':'stars_business'},inplace=True) 
business_keep.head(10)
```

```python
reviews_keep=reviews.drop(["funny","cool"], axis=1)
reviews_keep.rename(columns={'stars':'stars_review','text':'review_text','date':'date_review'},inplace=True) 
reviews_keep.head(10)
```

```python
user_keep=user.drop(["name","yelping_since","review_count","funny","cool","friends","compliment_more","compliment_profile","compliment_cute","compliment_list","compliment_note","compliment_plain","compliment_cool","compliment_funny","compliment_hot","compliment_writer","compliment_photos"], axis=1)
user_keep.head(10)
```

```python
tip_keep=tip.drop('date', axis=1)
tip_keep.rename(columns={'text':'review_tip','compliment_count':'compliment_count_tip'},inplace=True) 
tip_keep.head(10)
```

```python
business_review=pd.merge(business_keep, reviews_keep, how='inner',on="business_id")
business_review
```

```python
business_review_tip=pd.merge(business_review, tip_keep, how='outer',on=["business_id","user_id"])
business_review_tip
```

```python
business_review_tip_user=pd.merge(business_review_tip, user_keep, how='left',on="user_id")
business_review_tip_user
```

```python
business_review_tip_user.to_csv(index=True,path_or_buf="C:/Users/58479/Desktop/628/module3/Stat628/Stat628/merged_data.csv")
```
