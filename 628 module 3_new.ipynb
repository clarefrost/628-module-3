```python
#directory_wkn="C:/Users/wkn98/stat628/628-module-3"
import json
import pandas as pd
def readjson(filepath):
    file = open(filepath, 'r', encoding='utf-8')
    dat={}
    for i,line in enumerate(file,start=1):
        dat_new=json.loads(line)
        if i==1:
            dat=dat_new
            for key in dat :
                dat[key]=[dat[key]]
        else:
            for key in dat:
                dat[key].append(dat_new[key])
    file.close()
    data=pd.DataFrame(dat)
    return data
reviews=readjson("../M3Data/review.json")
#reviews=readjson("/Users/hallu/Desktop/628-module-3-master/review.json")
reviews.head(10)
```

## remove stop words

```python
def filter_stop_words(words):
    stops=set(stopwords.words('english'))
    words = [word for word in words if word.lower() not in stops]
    return words
```

```python
import string

def token_review(review):
    st1="".join(review)
    st1_token=nltk.word_tokenize(st1)
    st1_token=[''.join(i for i in s if i not in string.punctuation) for s in st1_token]
    st1_token=filter_stop_words(st1_token)
    st1_token=[s for s in st1_token if s]
    return st1_token
```

## PlotWordStar

```python
import matplotlib.pyplot as plt
import numpy as np
def PlotWordStar(words,source):
    data = source['stars_review']
    text = source['review_text']
    data = [int(s) for s in data]
    data = pd.DataFrame({"stars":data})
    for i in words:
        init = []
        for s in text:
            init.append(s.count(i))
        init = pd.DataFrame({i:init})
        data = pd.concat([data,init],axis=1)
    fre = [0,0,0,0,0]
    star = [0,0,0,0,0]
    for word in words:
        fre = [0,0,0,0,0]
        star = [0,0,0,0,0]
        for index, row in data.iterrows():
            fre[int(row["stars"]-1)] += int(row[word]>0)
            star[int(row["stars"]-1)] += 1
        r,c = data.shape
        fre = np.array(fre)/r/(5*np.array(star)/r)
        plt.bar(["1","2","3","4","5"],fre)
        plt.title(word)
```
